## Aggregation

Muchos procesos de crecimiento observados en la naturaleza son el resultado de la agregación (acreción?) de partículas o elementos pequeños. La forma en la que se agregan estos elementos puede ser aleatoria y sorprendentemente dar lugar a estructuras con geometrías regulares (copos de nieve, polvo estelar). 

El prmer proceso de agregación que vamos a simular es uno de difusión limitada (DLA, diffusion limited aggregation). Las partículas se mueven de forma aleatoria pero cuando están cerca entre sí se agregan. Los agregados de forma individual pueden, eventualmente, crecer. La simulación de las párticulas se puede hacer con __caminatas aleatorias__, definidas como una sucesión de pasos que varian de forma aleatoria entre cada paso e independiente de los pasos previos.  

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Establecer las condiciones iniciales
#-------Definir el espacio, #de iter y #de particulas
N = 128 #tamaño de la cuadrícula
max_iter = 100000 #número máximo de iteraciones temporales
n_walkers = 1000 #Número de caminantes aleatorios
#-------Definir pasos movimiento de la partícula
x_step = np.array([-1,0,1,0])
y_step = np.array([0,-1,0,1])
dx = np.array([-1,0,1,0,-1,1,1,-1])
dy = np.array([0,-1,0,1,-1,-1,1,1])
#
grid = np.zeros([N+2,N+2], dtype = 'int') #Make grid with plus 2 to avoid boundary condition problems
x = np.zeros(n_walkers, dtype = 'int')
y = np.zeros(n_walkers, dtype = 'int')
status = np.ones(n_walkers, dtype= 'int')
#----Asignar de forma aleatoria la posición de las partículas
for i in range(0, n_walkers):
    x[i] = np.random.randint(0, N-1)
    y[i] = np.random.randint(0, N-1)
    grid[x[i],y[i]] = 1
grid[N//2,N//2] = 2 #sticky initial central node
```

```python
#make the random walk 
iteration,n_glued = 0,0

while (n_glued < n_walkers) and (iteration < max_iter):
    for i in range(0, n_walkers):
        if status[i] == 1:
            ii = np.random.choice([0,1,2,3])
            x_new = x[i] + x_step[ii]
            y_new = y[i] + y_step[ii]
            x_new = (N + x_new) % N
            y_new = (N + y_new) % N
            grid[x_new,y_new] = 1
            grid[x[i],y[i]] = 0
            x[i], y[i] = x_new, y_new
            if 2 in grid[x[i] + dx[:], y[i]+dy[:]]:
                grid[x[i], y[i]] = 2
                status[i] = 2
                n_glued += 1
        #End of work on single walker
    #end of iteration with all walkers    
    iteration += 1
    print('iteration {0}, glued walkers {1}.'.format(iteration, n_glued))
#End of temporal loop
plt.imshow(grid, interpolation='nearest')
plt.show()
```
